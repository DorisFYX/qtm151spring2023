- This is the residual sum of squares (rss)
- Call the function "fn_rss"

$\qquad (b_0 = 50, b_1 = -6 )$ or <br>
$\qquad (b_0 = 50, b_1 = 0.6 )$ or <br>
$\qquad (b_0 = 50, b_1 = 6)$ <br>
produce the lowest value of rss?


````python
# Write your own code
def fn_rss(y,x,b0,b1):
    e2 =(y-b0-b1*x)**2
    return (e2)

y = fifa["Overall"]
x =fifa["Age"]

sum1 =sum (fn_rss(y,x,50,-6))

sum2=sum (fn_rss(y,x,50,0.6))
sum3 =sum (fn_rss(y,x,50,6))
print(sum1)
print(sum2)
print(sum3)

# The second one produces the lowest value of rss




````
