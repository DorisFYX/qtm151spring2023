(b) Recode a numeric colum

<font size = "3">

- Import the "data_raw/seasons.csv" file
- Recode the "year" column into a new column "year_brackets" <br>
with the following categories

$\qquad$ ``` ["1950-1970","1971-1990","1991-2005", "2006-2020", "2020-onwards"] ```

$\qquad$ HINT: Use the "pd.cut()" command.

````python
# Write your own code
season = pd.read_csv('/Users/yuxinfeng/Documents/GitHub/qtm151spring2023/Lecture 08/Quiz 3 - Section 4/data_raw/seasons.csv')

results   = season.replace(to_replace = ["year"],

                            value = ["year_bracktes"])

bins= [1949,1970, 1990, 2005, 2020,3000]
labels_rank = ["1950-1970","1971-1990","1991-2005", "2006-2020", "2021-onwards"]
results["year_brackets"] = pd.cut(results["year"],
                              bins = bins,
                              right = True,
                              labels = labels_rank)

````

(c) Aggregate and query

<font size = "3">

- Import the "data_raw/driver_standings.csv" file
- Subset the observations for 'raceId $\ge$ 99' using ```.query()```
- Obtain the mean of "points" by "driverId" <br> 
 using ``` .groupby().agg() ```
- Subset the drivers with a mean of points $\ge 40$
- Store the results of all these operations in a dataframe <br>
```driver_standings_agg ```

````python
# Write your own code

driver = pd.read_csv('/Users/yuxinfeng/Documents/GitHub/qtm151spring2023/Lecture 08/Quiz 3 - Section 4/data_raw/driver_standings.csv')
driver_standings= (driver.query("raceId>=99").groupby(["driverId"])
                      .agg(mean_points = ('points','mean') ))

driver_standings_agg = (driver_standings.query('mean_points >= 40'))

display(driver_standings_agg)
````

(d) Aggregate and sort

- Sort the teams in descending order using of <br>

````python
# Write your own code

driver = pd.read_csv('/Users/yuxinfeng/Documents/GitHub/qtm151spring2023/Lecture 08/Quiz 3 - Section 4/data_raw/driver_standings.csv')

driver_mean= (driver.groupby(["driverId"])
                      .agg(mean_points = ('points','mean') ).sort_values(by='mean_points',ascending =False))


display(driver_mean)
````
